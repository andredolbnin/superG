```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u(x, y):
    return y * np.sin(x)

x = np.linspace(0, 1, 3 + 1)
y = np.linspace(0, 2, 4 + 1)
X, Y = np.meshgrid(x, y)
print(X)
print(Y)
Z = u(X, Y)

fig = plt.figure(figsize = (10, 8))
ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, Z, cmap = 'viridis')
print(X.shape, Y.shape, Z.shape)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
def run():
    fileName = r"C:\Users\andre\Desktop\coords.csv"
    CSVData = open(fileName)
    arr1 = np.genfromtxt(CSVData, delimiter = ',')
    M = len(arr1)
    N = len(arr1[0])
    realX = np.linspace(0, 4, M)
    realY = np.linspace(0, 3, N)
    X1, Y1 = np.meshgrid(realX, realY)

    fig = plt.figure(figsize = (10, 8))
    ax = plt.axes(projection = '3d')

    arr1t = np.transpose(arr1)
    print(arr1t.shape)
    ax.plot_surface(X1, Y1, arr1t, cmap = 'viridis')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    #ax.view_init(90, 20)
    
    #-----------------------------------------------------
    
    def realF(x, y):
        return np.sqrt(4 + x * y)
    
    Z2 = realF(X1, Y1)
    fig = plt.figure(figsize = (10, 8))
    ax = plt.axes(projection = '3d')
    ax.plot_surface(X1, Y1, Z2, cmap = 'viridis')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
```
